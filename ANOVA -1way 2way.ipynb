```python
import pandas as pd
import scipy.stats as stats
from statsmodels.formula.api import ols
```

```python
data=pd.read_csv("preprocessedData.csv")
```

```python
data.head(10)
```

```python
data.isna().sum()
```

```python
data=pd.get_dummies(data,drop_first=True)
```

```python
data.drop(['sl_no'],axis=1,inplace=True)
```

```python
data.head(10)
```

# one way ANOVA


```python
#H0: There is no significance between salary and work experience specialisation.
#H1: There is significance between salary and work experience specialisation.
```

```python
data.shape
```

```python
data=data[['salary','workex_Yes']]
```

```python
import statsmodels.api as sm
```

```python
ano1=ols('salary ~ workex_Yes',data=data).fit()
oneway=sm.stats.anova_lm(ano1,typ=2)
```

```python
oneway
```

```python
#since p value is higher than 0.05 we accept H0, i,e There is no significance between salary and work experience specialisation.
```
